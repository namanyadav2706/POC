```python
df=df[[ 'WZ_ERCOT (BIDCLOSE_LOAD_FORECAST)',
 'WZ_Coast (BIDCLOSE_LOAD_FORECAST)',
 'WZ_SouthCentral (BIDCLOSE_LOAD_FORECAST)',
 'WZ_ERCOT (RTLOAD)',
 'WZ_Coast (RTLOAD)',
 'WZ_NorthCentral (BIDCLOSE_LOAD_FORECAST)',
 'WZ_SouthCentral (RTLOAD)',
 'WZ_Southern (RTLOAD)',
 'WZ_Southern (BIDCLOSE_LOAD_FORECAST)',
 'WZ_NorthCentral (RTLOAD)',
 'WZ_East (BIDCLOSE_LOAD_FORECAST)',
 'WZ_North (RTLOAD)',
 'WZ_East (RTLOAD)',
 'WZ_West (BIDCLOSE_LOAD_FORECAST)',
 'WZ_North (BIDCLOSE_LOAD_FORECAST)',
 'WZ_West (RTLOAD)',
 'WZ_FarWest (BIDCLOSE_LOAD_FORECAST)',
 'HB_NORTH (RTLMP)',
 'WZ_FarWest (RTLOAD)',
       'HB_NORTH (DALMP)']]
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
y = y.reshape(len(y),1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 3)
```

```python
from catboost import CatBoostRegressor
regressor = CatBoostRegressor()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(estimator = regressor, X = X_train, y = y_train,scoring='r2', cv = 10)
```
